```python
# -*- coding: utf-8 -*-
"""
Exemplo da Unidade 1, Seção 3: Princípios Fundamentais da Matemática e da Lógica
"""

import math
from itertools import permutations

print("--- Unidade 1, Seção 3: Princípios Fundamentais da Matemática e da Lógica ---")
print("Esta seção aborda conceitos de matemática discreta, combinatória, listas, arranjos, permutações, combinações e probabilidade, fundamentais para a lógica computacional [1, 2].\n")

# --- 1. Listas e o Princípio da Multiplicação ---
print("### 1. Listas e o Princípio da Multiplicação ###")
print("Uma lista é uma sequência ordenada de objetos, onde a ordem é significativa e pode conter elementos repetidos [5]. Uma lista de 'n' elementos é chamada de n-upla [6].")
print("O princípio da multiplicação é usado para contar listas: se há 'n' escolhas para o primeiro elemento e 'm' para o segundo, o total é n * m [7, 8].\n")

# Exemplo de formação de listas de comprimento dois com vogais e algarismos [7]
vogais = ['A', 'E', 'I', 'O', 'U']
algarismos = [9-17]

num_vogais = len(vogais) # n opções
num_algarismos = len(algarismos) # m opções

total_listas = num_vogais * num_algarismos
print(f"Número de vogais (n): {num_vogais}")
print(f"Número de algarismos (m): {num_algarismos}")
print(f"Total de listas de comprimento dois (vogal, algarismo) = n * m = {num_vogais} * {num_algarismos} = {total_listas}\n")

# --- 2. Fatorial, Permutações, Arranjos e Combinações ---
print("### 2. Fatorial, Permutações, Arranjos e Combinações ###")
print("Esses são diferentes tipos de agrupamentos (ordenados ou não) de elementos [18].\n")

# Fatorial (n!) [19]
print("--- Fatorial (n!) ---")
print("O fatorial de n (n!) é o produto de todos os inteiros positivos menores ou iguais a n, usado para contar permutações de n objetos distintos [19].")
n_fatorial = 5
resultado_fatorial = math.factorial(n_fatorial)
print(f"{n_fatorial}! = {resultado_fatorial}\n")

# Permutações (P_n = n!) [20, 21]
print("--- Permutações (ordenação de todos os elementos, a ordem importa) ---")
objetos = ['a', 'b', 'c'] # Exemplo: ordenar 3 objetos [19, 22]
num_objetos_perm = len(objetos)
num_permutacoes = math.factorial(num_objetos_perm) # P(n) = n!

print(f"Objetos: {objetos}")
print(f"Número de permutações (ordenações) de {num_objetos_perm} objetos: {num_permutacoes}")
print("As permutações são:")
for p in permutations(objetos):
    print(f"  {p}")
print("")

# Arranjos (A(n, p) = n! / (n-p)!, a ordem importa) [23]
print("--- Arranjos (seleção ordenada de p elementos de n, a ordem importa) ---")
elementos = [9-12] # Exemplo: arranjos de 4 elementos tomados 2 a 2 [23]
n_arranjos = len(elementos) # Total de elementos
p_arranjos = 2 # Elementos a serem selecionados

# math.perm disponível no Python 3.8+
num_arranjos = math.perm(n_arranjos, p_arranjos) # A(n, p) = n! / (n-p)!

print(f"Elementos: {elementos}")
print(f"Número de arranjos de {n_arranjos} elementos tomados {p_arranjos} a {p_arranjos}: {num_arranjos}")
print("Alguns arranjos (listas ordenadas):")
# Gerar e imprimir alguns arranjos para ilustração
count = 0
for arr in permutations(elementos, p_arranjos):
    if count < 5: # Imprime apenas os 5 primeiros para não ser muito longo
        print(f"  {arr}")
        count += 1
    else:
        break
if num_arranjos > 5:
    print("  ...")
print("")

# Combinações (C(n, p) = n! / (p! * (n-p)!), a ordem NÃO importa) [21, 24]
print("--- Combinações (seleção não ordenada de p elementos de n, a ordem NÃO importa) ---")
funcionarios = ['A', 'B', 'C', 'D', 'E'] # Exemplo: escolher 3 funcionários de 5 [24]
n_combinacoes = len(funcionarios) # Total de funcionários
p_combinacoes = 3 # Funcionários a serem promovidos

# math.comb disponível no Python 3.8+
num_combinacoes = math.comb(n_combinacoes, p_combinacoes) # C(n, p) = n! / (p! * (n-p)!)

print(f"Funcionários: {funcionarios}")
print(f"Número de combinações de {n_combinacoes} funcionários tomados {p_combinacoes} a {p_combinacoes}: {num_combinacoes}")
print("As combinações (subconjuntos) são:")
from itertools import combinations
for comb in combinations(funcionarios, p_combinacoes):
    print(f"  {comb}")
print("Note que a ordem não importa: (A,B,C) é o mesmo que (B,C,A) para combinações [25].\n")

# --- 3. Exemplo Prático: Probabilidade de Sobrecarga na Rede (Desafio da Unidade 1, Seção 3) ---
print("### 3. Exemplo Prático: Probabilidade de Sobrecarga na Rede ###")
print("Este é o desafio 'Sem medo de errar' da Seção 3 da Unidade 1 [4].")
print("Problema: Uma rede de computadores possui 3 escritórios (A, B, C), cada um com 5 pontos de acesso. Havia 13 computadores conectados e uma sobrecarga foi detectada em um escritório com 5 computadores. Qual a probabilidade de a sobrecarga ter vindo do escritório C? [4]\n")

num_escritorios = 3
max_pontos_por_escritorio = 5
total_conectados = 13
escritorio_alvo = 'C'
conexao_sobrecarga_escritorio = 5

# Lista para armazenar todas as distribuições válidas
todas_distribuicoes = []

# Gerar todas as combinações possíveis de conexões para os 3 escritórios [26, 27]
# Iterar sobre o número de conexões para cada escritório
for c_A in range(max_pontos_por_escritorio + 1):
    for c_B in range(max_pontos_por_escritorio + 1):
        for c_C in range(max_pontos_por_escritorio + 1):
            # Verificar se a soma total de conexões é 13 [26]
            if c_A + c_B + c_C == total_conectados:
                todas_distribuicoes.append({'A': c_A, 'B': c_B, 'C': c_C})

print(f"Total de distribuições válidas para {total_conectados} computadores conectados (máx. {max_pontos_por_escritorio} por escritório): {len(todas_distribuicoes)}")
print("As distribuições (conexões por escritório) são:")
for dist in todas_distribuicoes:
    print(f"  {dist}")
print("")

# Cenário (I) e (II) conforme a fonte [26, 27]
# Cenario I: (5, 5, 3) em alguma ordem
cenario_I_base = sorted([11, 13, 13])
cenario_I_distribuicoes = []
for perm in set(permutations(cenario_I_base)): # Usa set para evitar permutações idênticas
    if sum(perm) == total_conectados and all(x <= max_pontos_por_escritorio for x in perm):
        cenario_I_distribuicoes.append(perm)

# Cenario II: (5, 4, 4) em alguma ordem
cenario_II_base = sorted([12, 12, 13])
cenario_II_distribuicoes = []
for perm in set(permutations(cenario_II_base)): # Usa set para evitar permutações idênticas
    if sum(perm) == total_conectados and all(x <= max_pontos_por_escritorio for x in perm):
        cenario_II_distribuicoes.append(perm)

# A fonte lista as 6 possibilidades ordenadas, então vamos recriar a lista final de forma ordenada
# [26]: A=5, B=5, C=3; A=5, B=3, C=5; A=3, B=5, C=5
# [27]: A=5, B=4, C=4; A=4, B=5, C=4; A=4, B=4, C=5
todas_distribuicoes_ordenadas = [
    {'A': 5, 'B': 5, 'C': 3},
    {'A': 5, 'B': 3, 'C': 5},
    {'A': 3, 'B': 5, 'C': 5},
    {'A': 5, 'B': 4, 'C': 4},
    {'A': 4, 'B': 5, 'C': 4},
    {'A': 4, 'B': 4, 'C': 5},
]
print(f"Total de cenários distintos e ordenados para a distribuição das 13 conexões: {len(todas_distribuicoes_ordenadas)}")
print("Cenários:")
for i, dist in enumerate(todas_distribuicoes_ordenadas):
    print(f"  Cenário {i+1}: A={dist['A']}, B={dist['B']}, C={dist['C']}")


# Contar cenários favoráveis (onde o escritório C tem 5 computadores conectados) [27]
cenarios_favoraveis = []
for dist in todas_distribuicoes_ordenadas:
    if dist[escritorio_alvo] == conexao_sobrecarga_escritorio:
        cenarios_favoraveis.append(dist)

num_total_cenarios = len(todas_distribuicoes_ordenadas)
num_cenarios_favoraveis = len(cenarios_favoraveis)

probabilidade = num_cenarios_favoraveis / num_total_cenarios if num_total_cenarios > 0 else 0

print(f"\nNúmero total de cenários possíveis para a distribuição de computadores: {num_total_cenarios}")
print(f"Número de cenários favoráveis (Escritório {escritorio_alvo} com {conexao_sobrecarga_escritorio} conexões): {num_cenarios_favoraveis}")
print(f"Os cenários favoráveis são:")
for dist in cenarios_favoraveis:
    print(f"  A={dist['A']}, B={dist['B']}, C={dist['C']}")

print(f"\nProbabilidade de a sobrecarga ter partido do Escritório {escritorio_alvo}: {num_cenarios_favoraveis}/{num_total_cenarios} = {probabilidade:.2f} ({probabilidade*100:.0f}%) [27].")

print("\nConforme a fonte, esta análise revela as possibilidades, mas a interpretação de qual escritório 'causou' a sobrecarga é um detalhe que deve ser considerado na comunicação [27].")
```
